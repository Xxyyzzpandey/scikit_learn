```python
import numpy as np
import pandas as pd
X=pd.DataFrame(np.array([5,3,2,np.NaN,np.NaN,4,-3,2,1,8,np.NaN,6,4,3,np.NaN]).reshape(5,3))
X.columns=['f1','f2','f3']
print(X)
z=X
```

```python
X.isnull().sum()
```

```python
X.isnull().sum().sum()
```

```python
from sklearn.impute import MissingIndicator
indicator=MissingIndicator(missing_values=np.NaN)
indicator=indicator.fit_transform(X)
indicator=pd.DataFrame(indicator,columns=['a1','a2','a3'])
print(indicator)
```

```python
from sklearn.impute import SimpleImputer
# imp=SimpleImputer(missing_values=np.NaN,strategy='mean')
imp=SimpleImputer(missing_values=np.NaN,strategy='median')
X=pd.DataFrame(imp.fit_transform(X))
X.columns=['f1','f2','f3']
print(X)
```
